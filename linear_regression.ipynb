```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Step 1: Load the dataset
df = sns.load_dataset('tips')
X= df[['total_bill']]
y= df['tip']
```

```python
# Step 2: Create the model
model= LinearRegression()
```

```python
# Step 3: Fit the model (Train the model)
model.fit(X,y)
```

```python
# Step 4: Predict from model
model.predict([[112]])
```

```python
# Lets plot the model
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
# Lets fit the best fit line
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.plot(X, model.predict(X), color= 'red')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
model.get_params
```

```python
# slope, intercept, r2 value
print('slope :', model.coef_)
print('Intercept :', model.intercept_)
print('R Squared :', model.score(X,y))
```

# SLR on the scaled data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler
 
# Load the dataset
df= sns.load_dataset('tips')
X= df[['total_bill']]
y=df['tip']

# Scale X
scalar= MinMaxScaler()
X_scaled=scalar.fit(X)

# Transform data (1-D array to 2-D array)
X_scaled= scalar.transform(X)

# Create a model
model= LinearRegression()

# Train a model
model.fit(X_scaled, y)
```

```python
# slope, intercept, r2 value
print('slope :', model.coef_)
print('Intercept :', model.intercept_)
print('R Squared :', model.score(X_scaled,y))
```

```python
# Loading the dataset
df= sns.load_dataset('iris')

# importing the necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Splitting the dataset into X and y (Features and labels)
X= df[['sepal_length','sepal_width']]
y= df['petal_width']

# Training the model on the training set
model= LinearRegression()
model.fit(X,y)
```

```python
print('slope :', model.coef_)
print('Intercept :', model.intercept_)
```

```python
model.predict([[5.1,3.5]])
```

```python
# importing the necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# splitting dataset into X and y (features and labels)
X= df[['sepal_length','sepal_width']]
y- df['petal_width']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
# training the model on the training set
model=LinearRegression() # calling the model
model.fit(X_train,y_train) # fitting or training the model (fitting a regression line)

# predicting the values of testing dataset
y_pred= model.predict(X_test)
```

```python
print('slope :', model.coef_)
print('Intercept :', model.intercept_)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
# printing the metrices for evaluating the model
print('R squred: ', r2_score(y_test, y_pred))
print("Mean Square Error:", mean_squared_error(y_test, y_pred))
print("Mean Absolute Error: ", mean_absolute_error(y_test, y_pred))
```
